```python
# Show unique values for each variable 
def display_unique_values(dataframe):
    for column in dataframe.columns:
        unique_values = dataframe[column].unique()
        print(f"Unique values in column '{column}':\n {unique_values}\n")

display_unique_values(df_train)
```

```python
# Check for missing data
def display_nul_values(dataframe):
    nan_found = False
    for column in dataframe.columns:
        nul_values = dataframe[column].isna().sum()
        if nul_values > 0:
            print(f"Null values in column '{column}':\n {nul_values}\n")
            nan_found = True

    if not nan_found:
        print('No nan columns identified')

display_nul_values(df_train)
```

```python
# Chi-square : test of independence of variables is used to assess whether two categorical variables are related
# to each other. It's often used for feature selection with categorical features in a dataset. We define a
# threshold against which we will test our feature's p-value and decide whether to drop ones that failed it.

categorical_cols = ['temp_of_extremities', 'peripheral_pulse', 'mucous_membrane','capillary_refill_time','pain',
                    'peristalsis','abdominal_distention','nasogastric_tube','nasogastric_reflux','rectal_exam_feces',
                    'abdomen','abdomo_appearance','lesion_2','surgery', 'age', 'surgical_lesion', 'lesion_3',
                    'cp_data']
threshold = .05

print(f'{"Feature":<25} | Chi-Square Test Result')
print('----------------------------------------')

for column in categorical_cols:
    # Create a contingency table
    contingency_table = pd.crosstab(df_train[column], df_train['outcome'])
    
    # Perform the Chi-Square test
    chi2, p, _, _ = stats.chi2_contingency(contingency_table)
    
    print(f'{column:<25} |   ', '\033[32mPassed' if p < threshold else '\033[31mFailed', '\033[0m')
```

# Modelling

```python
# Check the shapes of the train & test datasets
print('df_train shape: ', df_train.shape)
print('df_test shape: ', df_test.shape)
```

```python
X_train = df_train.drop(columns='outcome')
y_train = df_train['outcome'].map({'died':0,'euthanized':1,'lived':2})
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler, QuantileTransformer, OrdinalEncoder

numerical_pipeline = make_pipeline(
#     SimpleImputer(strategy='mean'), # Tree based models like the LGBM deal with missing values better than SimpleImputer
    QuantileTransformer(output_distribution='normal', random_state=42),
    StandardScaler()
)

categorical_pipeline = make_pipeline(
#     SimpleImputer(strategy='most_frequent'),
    OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=10)
)
```

```python
from sklearn.compose import make_column_transformer, make_column_selector

transformer = make_column_transformer(
    (
        numerical_pipeline,
        make_column_selector(dtype_include=np.number) # We want to apply numerical_pipeline only on numerical columns
    ),
    (
        categorical_pipeline,
        make_column_selector(dtype_include=object) # We want to apply categorical_pipeline only on object (string) columns
    ),
    remainder='passthrough', # If any columns were missed then don't drop them - we take care of every column so this line is not necessery
    verbose_feature_names_out=False # if False transformer won't add prefixes (name of the transformer that generated specific feature) to column names, column names are shorter that way
)

transformer
```

```python
X_train = transformer.fit_transform(X_train)
```

```python
X_train = pd.DataFrame(data=X_train, columns=transformer.get_feature_names_out(), index=df_train.index)
X_train.head()
```

```python
y_train
```

```python
from sklearn.model_selection import train_test_split

X_train_s, X_val, y_train_s, y_val = train_test_split(X_train, y_train, train_size=0.85)
```

```python
def train_and_evaluate_model(X_train_s, y_train_s, model_type):
    """
    Trains a classification model, generates predictions, and calculates predictive equality.
    
    Parameters:
        X_train_scaled (array-like): Scaled training features.
        y_train (array-like): Training target values.
        model_type (str): Type of model to use (default is 'SGDClassifier').
        
    Returns:
        tuple: A tuple containing the trained model.
    """
    if model_type == 'SVC':
        from sklearn.svm import SVC
        model = SVC(kernel='rbf', C=5)
        
    if model_type == 'LogisticRegression':
        from sklearn.linear_model import LogisticRegression
        model = LogisticRegression(max_iter=1000)
        
    if model_type == 'RandomForestClassifier':
        from sklearn.ensemble import RandomForestClassifier
        model = RandomForestClassifier()
    
    if model_type == 'GradientBoostingClassifier':
        from sklearn.ensemble import GradientBoostingClassifier
        model = GradientBoostingClassifier()
    
    if model_type == 'LGBM':
        from lightgbm import LGBMClassifier
        model = LGBMClassifier()
    
    model.fit(X_train_s, y_train_s)
    
    return model, model_type
```

```python
from sklearn.metrics import ConfusionMatrixDisplay, classification_report, f1_score

def calculate_model_metrics(model, X_train_s, y_train_s):
    # Calculate y predicted (y_preds)
    y_preds = model.predict(X_train_s)

    # Calculate model score
    model_score = round(model.score(X_train_s, y_train_s) * 100, 2)
    
    # Calculate f1-score
    micro_f1 = round(f1_score(y_train_s, y_preds, average='micro'), 2)
    macro_f1 = round(f1_score(y_train_s, y_preds, average='macro'), 2)
    
    # Print metrics
    print('Model Name:', model_type)
    print('Model Score:', model_score, "%")
    print('Model micro average F1-score: ', micro_f1)
    print('Model macro average F1-score: ', macro_f1)
    print('')
    #print('                 Classification Report')
    #print(classification_report(y_train,y_preds))
    
    # Plot ROC Confusion Matrix Display
    ConfusionMatrixDisplay.from_predictions(y_train_s, y_preds)

    plt.show()
```

```python
# Train the model SVC model & obtain metrics
model, model_type = train_and_evaluate_model(X_train_s, y_train_s, model_type='SVC')
print(model_type, 'Model metrics')
calculate_model_metrics(model, X_train_s, y_train_s)
```

```python
print(model_type, 'Model validation metrics')
calculate_model_metrics(model, X_val, y_val)
```
